<pre>
<ol>

<li> Compute F1 Score </li>
<li> Compute AUC Score, you need to compute different thresholds and for each threshold compute tpr,fpr and then use               numpy.trapz(tpr_array, fpr_array) <a href='https://stackoverflow.com/q/53603376/4084039'>https://stackoverflow.com/q/53603376/4084039</a>, <a href='https://stackoverflow.com/a/39678975/4084039'>https://stackoverflow.com/a/39678975/4084039</a> Note: it should be numpy.trapz(tpr_array, fpr_array) not numpy.trapz(fpr_array, tpr_array)

```python
df_a=pd.read_csv('5_a.csv')
df_a.head(5)
```

```python
# write your code here for task A
def confusion_matrix(df,threashold):
    TP=0
    FP=0
    TN=0
    FN=0
    y_pred=[]
    for i in df['proba']:

        if i<threashold:
            y_pred.append(0)
        else:
            y_pred.append(1)
    df['pred']=y_pred
    for i,j in zip(df['y'],df['pred']):
        if i==1 & j==1:
            TP+=1
        elif i==1 & j==0:
            FN+=1
        elif i==0 & j==1:
            FP+=1
        else:
            TN+=1
    return TP,FP,TN,FN
           
```

```python
# threashold value is taken as 0.51 because at threshold 0.50 no records are found
TP,FP,TN,FN=confusion_matrix(df_a,0.51)
```

```python
#2.F1-score
PR = TP/(TP+FP)
RE = TP/(TP+FN)

F1_Score = 2*(PR*RE)/(PR+RE)
print(F1_Score)
```

```python
#3.AUC and ROC 
# TPR = TP/(TP+FN)
# FPR = FP/FP+TN

TPR=[]
FPR=[]
y_sort = sorted(df_a['proba'],reverse = True)

for i in range(len(y_sort)):
    TP,FP,TN,FN=confusion_matrix(df_a,y_sort[i])
    tpr = TP/(TP+FN)
    TPR.append(tpr)
    fpr = FP/(FP+TN)
    FPR.append(fpr)
    
print(np.trapz(TPR,FPR))   
```
